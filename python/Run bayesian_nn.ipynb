```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
datasets =['Boston Housing','Concrete', 'Energy','Kin8nm']
dataset = datasets[3]

rmse_string = './rmseResult_' + dataset + '.npy'
ll_string = './llResult_' + dataset + '.npy'
iter_string = './iterResult_' + dataset + '.npy'

subset_rmse_string = './subset_rmseResult_' + dataset + '.npy'
subset_ll_string = './subset_llResult_' + dataset + '.npy'
subset_iter_string = './subset_iterResult_' + dataset + '.npy'

subset_1adver_rmse_string = './subset_1adver_rmseResult_' + dataset + '.npy'
subset_1adver_ll_string = './subset_1adver_llResult_' + dataset + '.npy'
subset_1adver_iter_string = './subset_1adver_iterResult_' + dataset + '.npy'


rmse_results = np.load(subset_1adver_rmse_string)
ll_results = np.load(subset_1adver_ll_string)
iter_results = np.load(subset_1adver_iter_string)
```

```python
fig = plt.figure(figsize=(10, 4))
t = np.arange(0, 100, 5);
#tInd = range(10,30)
tInd = range(10,20)

legendEntries1 = ('SVGD', 'Random Subset', 'Random Subset + Control Functional')
#legendEntries = ('SVGD', 'Random Subset', 'Random Subset + Control Functional','Induced Points')
legendEntries = ('SVGD', 'Random Subset', 'Random Subset + Control Functional','Random Induced Points', 
                 'Adversarial Induced Points (Batch)')
#plotColors = ['ro','bs','bd','yP','yX']
plotColors = ['-o',':s',':d','--P','--X','-->']
numModels = len(legendEntries)

plt.set_cmap('Set1')

ax=plt.subplot(121)
for i in range(0, numModels):
    plt.plot(t[tInd], rmse_results[i,tInd], plotColors[i])
    
plt.xlabel('time (sec)')
plt.ylabel('RMSE')
plt.title('Root Mean Squared Error')


#plt.legend( lines,legendEntries,loc='upper right')


#plt.subplot(132)
#for i in range(0, numModels):
#    plt.plot(t[tInd], ll_results[i,tInd], plotColors[i])

#plt.title('Log Likelihood')
#plt.xlabel('time (sec)')
#plt.ylabel('llh')

plt.subplot(122)
for i in range(0, numModels):
    plt.plot(t[tInd], iter_results[i,tInd], plotColors[i])

plt.xlabel('time (sec)')
plt.ylabel('iterations')
plt.title('Reached Iterations')
plt.legend(legendEntries, loc=9, bbox_to_anchor=(-0.1, -0.2),ncol = 3)

#plt.legend( legendEntries,loc='upper right')
plt.tight_layout()

plt.show()




```
